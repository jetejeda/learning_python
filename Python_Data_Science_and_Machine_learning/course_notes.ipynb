# Matplotlib
You can use 
```python
import matplotlib.pyplot as plt
%matplotlib inline
```
in your notebooks to get inline visualizations without using the .show() method. If you add a ';' at the end of the plot, the engine will only display the visualization, if you don't add it, it will first show you data related to your plot

## Plot components
- Figure: It's what holds our visualizations. **They can contain several axes** you can customize it using:
```python
plt.figure(facecolor="orange")# use it before calling plot
plt.plot(data)
```
- Axes: Where your graphic is actually created. It includes everything related to the visualization (name, labels, bars, etc.)

**It's a best practice to start creating your figure and main axes**
## Create several axes in a single figure
```python
plt.subplots(ncols=2, nrows=2)#It will create a figure with 4 axes
```

## create histogram using matplotlib
You can define the amount of groups your data will have (when working with an array of a single axis), this will be represented as the amount of stacked bars in your plot.

With alpha you can define the opacity for your data and the color param that you want for your histogram


````python
import numpy as np
data = np.random.randn(1000)
plt.hist(data, bins=30, alpha=0.8, color="red", edgecolor="white");
````

````python
x1 = np.random.randn(1000)
x2 = np.random.randn(500)
x3 = np.random.randn(100)
plt.hist(x1, histtype="stepfilled", alpha=0.3, bins=40);
plt.hist(x2, histtype="stepfilled", alpha=0.3, bins=40);
plt.hist(x3, histtype="stepfilled", alpha=0.3, bins=40);
````

## Scatter plots
You can create one by using the same syntax as the line plot but define a custom pattern that only includes specifications for the data points. You can also use the plt.scatter() method, with it, we gain customization capabilities.

````python
x = np.linspace(0,10,30)# Create 30 numbers from 0 to 10
y = np.sin(x)
````

````python
plt.plot(x,y, 'o');
````

````python
plt.scatter(x,y);
````

````python
random_generator = np.random.RandomState(0)
plt.scatter(random_generator.randn(100),
            random_generator.randn(100),
            c=random_generator.rand(100),
            s=random_generator.randn(100)*1000,
            alpha=0.6,
            cmap="viridis",)
plt.colorbar();
````

## Create multiple plots

As mentioned, you can use subplots method to achieve this. But how do we create different plots for each axes?

To do so, we can use the axs object. It's an array so you can navigate to specific axes using indexes.

````python
fig, axs = plt.subplots(ncols=2)
axs1 = axs[0]
axs2 = axs[1]

plt.style.use('ggplot')
axs1.set_title("Line plot")
axs2.set_title("Scatter plot")
axs1.plot([1,2,3,4], [1,4,9,16])
axs2.scatter([1,2,3,4], [1,4,9,16]);
````

# Common style for all your visualizations
You can use the plt.style.use() method to set the style for all your visualizations. With this, you don't have to specify the same attributes every time you want a new visualization with the same style.

````python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.DataFrame({
    "frutas": ['Manzanas', 'Bananas', 'Cerezas', 'Duraznos'],
    "cantidades": [40, 25, 15, 20]
})

plt.style.use("dark_background")
plt.pie(df["cantidades"],labels=df["frutas"],autopct="%1.1f%%",explode=(0.1,0.1,0.1,0.1))
plt.title("Distribución de frutas en el almacén")
````

````python
import matplotlib.pyplot as plt
import numpy as np


# make data
x = np.arange(1,11)
y = x**2

fig,ax = plt.subplots()
ax.stem(x, y)
ax.set_xlabel("Número (x)")
ax.set_ylabel("Cuadrado del número (x^2)")
ax.set_title("Relación entre Números y sus Cuadrados")
````

````python
import matplotlib.pyplot as plt
import numpy as np


# make data
x = np.arange(1,6)
y1= [6, 1, 4, 4, 8]
y2= [4, 6, 3, 5, 8]
y3= [7, 9, 9, 2, 7]


# plot
fig, ax = plt.subplots()

ax.stackplot(x,
y1,y2,y3,
labels=["Serie y1", "Serie y2", "Serie y3"], 
colors=["red", "blue", "green"])

ax.set_xlabel("Tiempo")
ax.set_ylabel("Valores Acumulados")
ax.set_title("Distribución acumulativa de valores por serie de tiempo")
ax.legend(loc="upper left")
````

````python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)
data1 = np.random.normal(loc=0, scale=0.5, size=100)
data2 = np.random.normal(loc=2, scale=1, size=100)
data3 = np.random.normal(loc=-2, scale=1.5, size=100)

fig, ax = plt.subplots()

ax.set_title("Comparación de distribuciones en Violin Plot")
vp = ax.violinplot([data1, data2, data3], positions=[1, 2, 3])
ax.set_xticks([1, 2, 3])
ax.set_xticklabels(["Muestra 1", "Muestra 2", "Muestra 3"])
````
